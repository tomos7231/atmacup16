```python
import pandas as pd
from pathlib import Path

class CFG:
    name = "ens002"
    seed = 42

    path_input = Path("../input")
    path_output = Path("../output")
```

```python
def setup(CFG):
    # expフォルダを作成
    CFG.path_exp = CFG.path_output / CFG.name
    CFG.path_exp.mkdir(parents=True, exist_ok=True)

    return CFG

CFG = setup(CFG)
```

```python
sub1 = pd.read_csv("../output/rule001/submission.csv")
sub2 = pd.read_csv("../output/exp018/submission.csv")

display(sub1.head())

sub1 = sub1.values
sub2 = sub2.values
```

```python
# sub1の各行で繰り返し
for i in range(sub1.shape[0]):
    # 対応する行を取得
    sub1_row = sub1[i]
    for j, pred in enumerate(sub1_row):
        # 予測値が0なら
        if pred == 0:
            for pred2 in sub2[i]:
                # sub1の予測値にないものを探す
                if pred2 not in set(sub1[i]):
                    sub1[i][j] = pred2
```

```python
sub = pd.DataFrame(sub1, columns=["predict_0", "predict_1", "predict_2", "predict_3", "predict_4", "predict_5", "predict_6", "predict_7", "predict_8", "predict_9"])
```
